```python
x=np.linspace(0, 14, 100)
y1 = x ** 2
c1, c2 = sns.color_palette("deep", 2)[0], sns.color_palette("deep", 2)[1]
plt.fill_between(x, y1 - 20, y1 + 20, color=c2, alpha=.2)
plt.plot(x, y1)
```

```python
from matplotlib.widgets import Slider

def plotShit(k):
    x = np.linspace(0, 14, 100)
    y1 = np.sin(k * x + .5)
    y2 = np.sin(k * x + 4 * .5) * 3
    c1, c2 = sns.color_palette("deep", 2)[0], sns.color_palette("deep", 2)[1]
    plt.plot(x, y1)
    plt.fill_between(x, y1 - .5, y1 + .5, color=c1, alpha=.2)
    plt.plot(x, y2)
    plt.fill_between(x, y2 - .8, y2 + .8, color=c2, alpha=.2);
    
slider_ax = plt.axes([0.1, 0.1, 0.8, 0.02])
slider = Slider(slider_ax, "Offset", -5, 5, valinit=0, color='#AAAAAA')
slider.on_changed(plotShit)
```

```python
fig, ax = plt.subplots()
fig.subplots_adjust(bottom=0.2, left=0.1)

t = np.linspace(0, 10, 1000)
line, = plt.plot(t, np.sin(t), lw=2)

def on_change(val):
    line.set_ydata(np.sin(t - val))

slider_ax = plt.axes([0.1, 0.1, 0.8, 0.02])
slider = Slider(slider_ax, "Offset", -5, 5, valinit=0, color='#AAAAAA')
slider.on_changed(on_change)
```
